```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file1 = '/content/drive/MyDrive/Colab Notebooks/Dress/Attribute+DataSet.csv'

data = pd.read_csv(file1)
```

```python
data.head()
```

```python
# Checking for data types
data.info()
```

```python
# Checking for missing values and data types in both datasets
data.isnull().sum()
```

```python
# Summary statistics for numerical columns
data.describe()
```

```python
#Count of Unique Values for Each Column
print("Unique Values:\n", data.nunique())
```

```python
# Handle missing data (optional, for cleaner analysis)
data.fillna("Unknown", inplace=True)

# Set a general style for visualizations
sns.set(style="whitegrid")
```

# **Univariate Analysis**

```python
# 1. Analyze the distribution of ratings
plt.figure(figsize=(8, 5))
sns.histplot(data['Rating'], kde=True, bins=10, color='blue')
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Frequency distribution of styles
plt.figure(figsize=(10, 6))
data['Style'].value_counts().plot(kind='bar', color='purple')
plt.title("Frequency Distribution of Styles")
plt.xlabel("Style")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Proportion of recommendations
plt.figure(figsize=(6, 6))
data['Recommendation'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=["Not Recommended", "Recommended"], colors=['red', 'green'])
plt.title("Proportion of Recommendations")
plt.show()
```

```python
# 4. Distribution of dress sizes
plt.figure(figsize=(8, 5))
sns.countplot(x=data['Size'], palette='viridis')
plt.title("Distribution of Dress Sizes")
plt.xlabel("Size")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Count of each price category
plt.figure(figsize=(8, 5))
data['Price'].value_counts().plot(kind='bar', color='orange')
plt.title("Count of Each Price Category")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# 6. Count of dresses by season
plt.figure(figsize=(8, 5))
data['Season'].value_counts().plot(kind='bar', color='teal')
plt.title("Count of Dresses by Season")
plt.xlabel("Season")
plt.ylabel("Count")
plt.show()
```

```python
# 7. Analysis of missing values in Material
missing_materials = data['Material'].value_counts()
plt.figure(figsize=(8, 5))
missing_materials.plot(kind='bar', color='pink')
plt.title("Distribution of Material")
plt.xlabel("Material")
plt.ylabel("Count")
plt.show()
```

```python
# 8. Analyze the occurrence of different necklines
plt.figure(figsize=(10, 6))
data['NeckLine'].value_counts().plot(kind='bar', color='cyan')
plt.title("Occurrence of Different Necklines")
plt.xlabel("NeckLine")
plt.ylabel("Count")
plt.show()
```

```python
# 9. Identify the most common fabric types
plt.figure(figsize=(10, 6))
data['FabricType'].value_counts().plot(kind='bar', color='lime')
plt.title("Most Common Fabric Types")
plt.xlabel("Fabric Type")
plt.ylabel("Count")
plt.show()
```

```python
# 10. Count of unique dress IDs
unique_dress_ids = data['Dress_ID'].nunique()
print(f"Number of unique dress IDs: {unique_dress_ids}")
```

# **Bivariate Analysis**

```python
# 1. Relationship between price and recommendation
plt.figure(figsize=(8, 5))
sns.countplot(x=data['Price'], hue=data['Recommendation'], palette='viridis')
plt.title("Relationship between Price and Recommendation")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Ratings by season
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['Season'], y=data['Rating'], palette='muted')
plt.title("Ratings by Season")
plt.xlabel("Season")
plt.ylabel("Rating")
plt.show()
```

```python
# 3. Average rating for each style
average_rating_style = data.groupby('Style')['Rating'].mean().sort_values()
plt.figure(figsize=(10, 6))
average_rating_style.plot(kind='bar', color='red')
plt.title("Average Rating for Each Style")
plt.xlabel("Style")
plt.ylabel("Average Rating")
plt.show()
```

```python
# 4. Size distribution across price categories
plt.figure(figsize=(8, 5))
sns.countplot(x=data['Size'], hue=data['Price'], palette='viridis')
plt.title("Size Distribution across Price Categories")
plt.xlabel("Size")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Effect of neckline type on recommendation
plt.figure(figsize=(10, 6))
sns.countplot(x=data['NeckLine'], hue=data['Recommendation'], palette='husl')
plt.title("Effect of Neckline Type on Recommendation")
plt.xlabel("NeckLine")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate average ratings for each season
average_ratings = data.groupby('Season')['Rating'].mean()

# Create a line plot
plt.figure(figsize=(8, 5))
plt.plot(average_ratings.index, average_ratings.values, marker='o', linestyle='-', color='orange')
plt.title("Average Ratings by Season")
plt.xlabel("Season")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add a grid for better visualization
plt.show()
```

```python
# 7. Correlation between ratings and recommendations
correlation = data[['Rating', 'Recommendation']].corr()
print("Correlation between Ratings and Recommendations:\n", correlation)
```

```python
# 8. Material vs. Recommendation distribution
plt.figure(figsize=(10, 6))
sns.countplot(x=data['Material'], hue=data['Recommendation'], palette='Set1')
plt.title("Material vs. Recommendation Distribution")
plt.xlabel("Material")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Decoration vs. Recommendation counts
plt.figure(figsize=(10, 6))
sns.countplot(x=data['Decoration'], hue=data['Recommendation'], palette='Dark2')
plt.title("Decoration vs. Recommendation Counts")
plt.xlabel("Decoration")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Pattern type vs. Recommendation analysis
plt.figure(figsize=(10, 6))
sns.countplot(x=data['Pattern Type'], hue=data['Recommendation'], palette='viridis')
plt.title("Pattern Type vs. Recommendation Analysis")
plt.xlabel("Pattern Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. Neckline vs. Average rating
average_rating_neckline = data.groupby('NeckLine')['Rating'].mean().sort_values()
plt.figure(figsize=(10, 6))
average_rating_neckline.plot(kind='bar', color='skyblue')
plt.title("Average Rating by Neckline")
plt.xlabel("NeckLine")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# 12. Sleeve length vs. Recommendation
plt.figure(figsize=(10, 6))
sns.countplot(x=data['SleeveLength'], hue=data['Recommendation'], palette='viridis')
plt.title("Sleeve Length vs. Recommendation")
plt.xlabel("Sleeve Length")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group data by Season and FabricType, then count occurrences
season_fabric_counts = data.groupby(['Season', 'FabricType'])['Dress_ID'].count().unstack()

# Create the stacked bar chart
season_fabric_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Customize the plot
plt.title('Season vs. Fabric Type Counts (Stacked)')
plt.xlabel('Season')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Fabric Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 14. Style vs. Recommendation proportions
style_recommendation = pd.crosstab(data['Style'], data['Recommendation'], normalize='index')
style_recommendation.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title("Style vs. Recommendation Proportions")
plt.xlabel("Style")
plt.ylabel("Proportion")
plt.show()
```

```python
# 15. Price vs. Ratings
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['Price'], y=data['Rating'], palette='pastel')
plt.title("Price vs. Ratings")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

# **Multivariate Analysis**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Select only numerical columns for the heatmap
numeric_data = data.select_dtypes(include=np.number)

# Create a heatmap with the 'viridis' colormap
sns.heatmap(numeric_data.corr(), cmap='viridis', annot=True)  # Add annot=True to display correlation values
plt.title('Correlation Matrix of Numerical Features')  # Add a title for better understanding
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a pivot table to aggregate ratings by size and season
rating_pivot = pd.pivot_table(data, values='Rating', index='Size', columns='Season')

# Generate the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(rating_pivot, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title("Average Ratings across Combinations of Size and Season")
plt.xlabel("Season")
plt.ylabel("Size")
plt.show()
```
